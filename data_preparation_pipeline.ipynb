```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/hr_workforce_clean.csv')
df.head()
```

```python
df['employee_id'] = range(1, len(df) + 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['birthdate'] = pd.to_datetime(df['birthdate'], dayfirst=True, errors='coerce')
df['hiredate'] = pd.to_datetime(df['hiredate'], dayfirst=True, errors='coerce')
df['termdate'] = pd.to_datetime(df['termdate'], dayfirst=True, errors='coerce')

df.info()
```

```python
today = pd.to_datetime("today")

df['age'] = ((today - df['birthdate']).dt.days / 365).round(1)
```

```python
df['end_date'] = df['termdate'].fillna(today)

df['tenure_years'] = ((df['end_date'] - df['hiredate']).dt.days / 365).round(1)
```

```python
df['employment_status'] = df['termdate'].apply(
    lambda x: 'Active' if pd.isna(x) else 'Terminated'
)
```

```python
df['gender'] = df['gender'].replace({
    'M': 'Male',
    'F': 'Female'
})
```

```python
df[['age', 'tenure_years', 'employment_status', 'gender']].head(10)
```

```python
# حذف الأعمدة المؤقتة
df.drop(columns=['end_date'], inplace=True)

# ترتيب الأعمدة (اختياري لكنه احترافي)
cols = [
    'employee_id', 'first_name', 'last_name', 'gender',
    'age', 'education_level', 'department',
    'state', 'city',
    'hiredate', 'termdate',
    'tenure_years', 'employment_status'
]

df = df[cols]
```

```python
df.to_csv('/content/hr_final_clean.csv', index=False)
```
